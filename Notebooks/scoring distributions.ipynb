```python
import pandas as pd
import numpy as np

import plotly.express as px
```

```python
data_path = '../Data/imdb_clean.csv'
df = pd.read_csv(data_path)

df.head()
```

```python
df['imdb_score'].plot.hist()
```

```python
df['movie_facebook_likes'].plot.hist()
```

```python
df['gross'].hist()
```

```python
df['imdb_score_uniform'] = pd.Series(df['imdb_score']).replace(np.sort(df['imdb_score']),np.linspace(0,1,len(df['imdb_score'])))
df['movie_facebook_likes_uniform'] = pd.Series(df['movie_facebook_likes']).replace(np.sort(df['movie_facebook_likes']),np.linspace(0,1,len(df['movie_facebook_likes'])))
df['budget_uniform'] = pd.Series(df['budget']).replace(np.sort(df['budget']),np.linspace(0,1,len(df['budget'])))
df['gross_uniform'] = pd.Series(df['gross']).replace(np.sort(df['gross']),np.linspace(0,1,len(df['gross'])))

df.plot(x='imdb_score_uniform', 
        y='movie_facebook_likes_uniform',
        c='budget_uniform',
        kind='scatter')
```

```python
fig = px.scatter_3d(
    df,
    x='imdb_score_uniform',
    y='movie_facebook_likes_uniform',
    z='budget_uniform',
    color='gross_uniform',
    opacity=0.6
)

fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
display(fig)
```
